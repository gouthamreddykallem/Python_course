##Question #2

Below code reads the image and converts it to array using NumPy library.

Perform the following actions.

    1 - read two images and reshape 120,120, lets call the images as image1 and image2  - 0 Points
    2 - Create an array of 1's same size as your image1 and add it to image1 - 2.5 points
    3- create an array of 0's of same size as your image2 and add it to image 2 - 2.5 points
    4 - concatinate both images - 2.5 Ponts
    5 - devide the step 4 image by 255 - 2.5 points
    


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

image1 = cv2.imread('Images/1.jpg')
image2 = cv2.imread('Images/2.jpg')
# 1
resize1 = cv2.resize(image1, dsize=(120,120))
resize2 = cv2.resize(image2, dsize=(120,120))

# 2 and 3
ones = np.arange(120*120*3).reshape(resize1.shape)
zeros = np.arange(120*120*3).reshape(resize1.shape)
ones.fill(1)
zeros.fill(0)
sum1 = (resize1) + (ones)
sum2 = (resize2) + (zeros)
# 4
combinedImage = cv2.vconcat([sum1, sum2])
plt.imshow(combinedImage)
```

```python
# 5
combinedImage = combinedImage/255
plt.imshow(combinedImage)
```

##Question # 3 
Points :10

    1 - Slice the image into 50x50
    2 - save the image and load the image


```python
tiles = [combinedImage[x:x+50,y:y+50] for x in range(0,combinedImage.shape[0],50) for y in range(0,combinedImage.shape[1],50)]
# plt.imshow(tiles[10])
path = "res_images"
for i in range(len(tiles)):
    slice_ = tiles[i]
    processedSlice = 255 * (slice_ - slice_.min()) / (slice_.max() - slice_.min())
    processedSlice = np.array(processedSlice, np.int)
    fileName = str(i)+".png"
    cv2.imwrite(os.path.join(path , fileName) ,processedSlice)

for f in os.listdir(path):
    image = cv2.imread(path+'/'+f)
    plt.figure()
    plt.imshow(image)
    plt.title(f)
```
